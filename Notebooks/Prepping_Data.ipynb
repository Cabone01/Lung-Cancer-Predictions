```python
import pandas as pd
```

```python
df = pd.read_csv("../Datasets/cancer_patient_datasets.csv")
df.head()
```

```python
df_clean = df.drop(columns=["index", "Patient Id"])
df_clean.head()
```

```python
df_clean.dtypes
```

```python
cancer_level = {"Low": 0, "Medium": 1, "High": 2}
df_prep = df_clean.replace({"Level": cancer_level})
df_prep.head()
```

```python
column_moved = df_prep.pop("Level")
df_prep.insert(0, "Cancer_Level", column_moved)
df_prep.head()
```

```python
df_prep.nunique(axis=0)
```

```python
df_prep.to_csv("../Datasets/Lung_Cancer_Clean_Data.csv", index = False)
```
